```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Đọc dữ liệu
df = pd.read_csv("data/train.csv")

# Tổng quan dữ liệu
print("Kích thước dữ liệu:", df.shape)
display(df.head())
```

```python
# Kiểm tra thiếu dữ liệu
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print("Các cột thiếu dữ liệu:")
display(missing)
```

```python
# Biểu đồ phân phối giá nhà
plt.figure(figsize=(10, 5))
sns.histplot(df["SalePrice"], kde=True, bins=40)
plt.title("Phân phối SalePrice")
plt.xlabel("SalePrice")
plt.show()
```

```python

# Tương quan với SalePrice
plt.figure(figsize=(12, 8))
corr = df.corr(numeric_only=True)["SalePrice"].sort_values(ascending=False)
sns.barplot(x=corr[1:11].values, y=corr[1:11].index)
plt.title("Top 10 đặc trưng tương quan cao nhất với SalePrice")
plt.show()
```
